```python
# importing the libraries  
import numpy as np
import pandas as pd
```

```python
# Reading the data
# this csv file is having only (cast crew and movie id) data till 2017 
credits = pd.read_csv('credits.csv')
```

```python
credits
```

```python
# this csv file is having other movies related features data till 2017 
meta = pd.read_csv('tmdb_5000_movies.csv')
```

```python
meta.head()
```

```python
# changing the release date from object to date time format
meta['release_date'] = pd.to_datetime(meta['release_date'], errors='coerce')
```

```python
# changing the release year from object to date year format
meta['year'] = meta['release_date'].dt.year
```

```python
# Counting the released number of movies in each year
meta['year'].value_counts().sort_index()
```

```python
# Getting only 2017 movies, as we already have movies up to the year 2016 in preprocessing_1
new_meta = meta[['id','title', 'genres', 'keywords','year', 'popularity', 'vote_average', 'vote_count']]
```

```python
new_meta
```

```python
new_meta.dropna(inplace=True)
```

```python
new_meta['id'] = new_meta['id'].astype(int)
```

```python
# merging the credits.csv data with movies_metadata.csv data on the basis on movie ID
# we are doing so , because both the files are having imformative data
data = pd.merge(new_meta, credits, on='id')
```

```python
pd.set_option('display.max_colwidth', 75)
data
```

```python
# evaluates an expression node or a string containing a Python literal or container display
import ast
data['keywords'] = data['keywords'].map(lambda x: ast.literal_eval(x))
data['genres'] = data['genres'].map(lambda x: ast.literal_eval(x))
data['cast'] = data['cast'].map(lambda x: ast.literal_eval(x))
data['crew'] = data['crew'].map(lambda x: ast.literal_eval(x))
```

```python
# Extracting the geners_list using this function 
def make_genresList(x):
    gen = []
    st = " "
    for i in x:
        if i.get('name') == 'Science Fiction':      # converting the "Science Fiction" to "Sci-Fi"
            scifi = 'Sci-Fi'
            gen.append(scifi)
        else:
            gen.append(i.get('name'))
    if gen == []:
        return np.NaN
    else:
        return (st.join(gen))
```

```python
data['genres_list'] = data['genres'].map(lambda x: make_genresList(x))
```

```python
data['genres_list']
```

```python
# Getting the cast information (First Cast)
def get_actor1(x):
    casts = []
    for i in x:
        casts.append(i.get('name'))
    if casts == []:
        return np.NaN
    else:
        return (casts[0])
```

```python
data['actor_1_name'] = data['cast'].map(lambda x: get_actor1(x))
```

```python
# Getting the cast information (Second Cast)
def get_actor2(x):
    casts = []
    for i in x:
        casts.append(i.get('name'))
    if casts == [] or len(casts)<=1:
        return np.NaN
    else:
        return (casts[1])
```

```python
data['actor_2_name'] = data['cast'].map(lambda x: get_actor2(x))
```

```python
data['actor_2_name']
```

```python
# Getting the cast information (Third Cast)
def get_actor3(x):
    casts = []
    for i in x:
        casts.append(i.get('name'))
    if casts == [] or len(casts)<=2:
        return np.NaN
    else:
        return (casts[2])
```

```python
data['actor_3_name'] = data['cast'].map(lambda x: get_actor3(x))
```

```python
data['actor_3_name']
```

```python
# getting the Director name for Each Movie by using this function
def get_directors(x):
    dt = []
    st = " "
    for i in x:
        if i.get('job') == 'Director':
            dt.append(i.get('name'))
    if dt == []:
        return np.NaN
    else:
        return (st.join(dt))
```

```python
data['director_name'] = data['crew'].map(lambda x: get_directors(x))
```

```python
data['director_name']
```

```python
# getting the keywords for Each Movie by using this function
def make_keywords_List(x):
    key = []
    st = " "
    for i in x:
        key.append(i.get('name'))
    if key == []:
        return np.NaN
    else:
        return (st.join(key))
```

```python
data['keywords_list'] = data['keywords'].map(lambda x: make_keywords_List(x))
```

```python
data['keywords_list']
```

```python
# making a new movie data , that is haing only this important information that we have extracted
movie = data.loc[:,['id','keywords_list','director_name','actor_1_name','actor_2_name','actor_3_name','genres_list','popularity','vote_average','vote_count','title','year']]
```

```python
movie
```

```python
movie.isna().sum()
```

```python
movie = movie.dropna(how='any')
```

```python
movie.isna().sum()
```

```python
# renaming the columns
movie = movie.rename(columns={'genres_list':'genres'})
movie = movie.rename(columns={'title':'movie_title'})
movie = movie.rename(columns={'keywords_list':'keywords'})
```

```python
movie['movie_title'] = movie['movie_title'].str.lower()
```

```python
# making a new column and Concatenate all the value into that single column 
movie['comb'] = movie['actor_1_name'] + ' ' + movie['actor_2_name'] + ' '+ movie['actor_3_name'] + ' '+ movie['director_name'] +' ' + movie['keywords']  + movie['genres'] + movie['popularity'].map(str)
```

```python
# saving the data into a csv file
movie.to_csv('hollywood_movie.csv',index=False)
```
